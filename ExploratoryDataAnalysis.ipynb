**1.	What data domain you have selected?**

`Retail and E-commerce Analytics:` the purpose of this is is to support business intelligence through analysis of sales trends, product performance, and customer purchase behaviour for Amazon (a retail e-commerce business)


**2. What does the dataset contain?**

`index` – A numerical index representing the row number in the dataset.

`Order ID` – A unique identifier for each order.

`Date` – The date the order was placed, formatted as MM-DD-YY.

`Status` – The current status of the order (e.g., Cancelled, Shipped).

`Fulfilment` – Indicates whether the order was fulfilled by the merchant or by Amazon.

`Sales Channel` – The platform through which the order was placed (e.g., Amazon.in).

`ship-service-level` – The shipping service level chosen for the order (e.g., Standard, Expedited).

`Style` – The product style or type identifier.

`SKU` – The Stock Keeping Unit, a unique identifier for each product variation.

`Category` – The category of the product (e.g., Set, Kurta, Western Dress).

`currency` – The currency in which the transaction occurred (e.g., INR).

`Amount` – The monetary value of the transaction.

`ship-city` – The city to which the product was shipped.

`ship-state` – The state to which the product was shipped.

`ship-postal-code` – The postal code for the shipping address.

`ship-country` – The country to which the product was shipped.

`promotion-ids` – Identifiers for promotions applied to the order.

`B2B` – Indicates whether the transaction was a business-to-business (B2B) order (True/False).

`fulfilled-by` – Indicates the fulfillment method (e.g., Easy Ship).

`Unnamed: 22` – An unutilized or unnamed column in the dataset (contains only 'NaN' and 'False' values).

***B.) For selected dataset generate Five Number Summary using Python. Also generate mode and midrange, outlier detection using concept of Quartile method and other. Compare the results.***

```python
# Five Number Summary
five_number_summary = data1[numerical_columns].describe().loc[['min', '25%', '50%', '75%', 'max']]
print("Five Number Summary:")
print(five_number_summary)

# Mode
mode_values = data1[numerical_columns].mode().loc[0]
print("\nMode:")
print(mode_values)

# Midrange
midrange_values = (data1[numerical_columns].min() + data1[numerical_columns].max()) / 2
print("\nMidrange:")
print(midrange_values)

# Outlier detection using Quartile method
Q1 = data1[numerical_columns].quantile(0.25)
Q3 = data1[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

outliers = ((data1[numerical_columns] < (Q1 - 1.5 * IQR)) | (data1[numerical_columns] > (Q3 + 1.5 * IQR)))
print("\nOutliers detected using Quartile method:")
print(outliers)

# Compare the results
comparison = pd.DataFrame({
    'Five Number Summary': five_number_summary.stack(),
    'Mode': mode_values.repeat(5).values,
    'Midrange': midrange_values.repeat(5).values
}, index=pd.MultiIndex.from_product([numerical_columns, ['min', '25%', '50%', '75%', 'max']]))
print("\nComparison of Five Number Summary, Mode, and Midrange:")
print(comparison.unstack())
```

```python
# Drop duplicate labels in the index
data1 = data1[~data1.index.duplicated(keep='first')]

numeric_columns = data1.select_dtypes(include=['float64', 'int64'])

def five_num_summary(numeric_columns):
    summaries = {}
    for column in numeric_columns:
        q1 = numeric_columns[column].quantile(0.25)
        q2 = numeric_columns[column].quantile(0.5)
        q3 = numeric_columns[column].quantile(0.75)
        iqr = q3 - q1
        lower_whisker = q1 - (1.5 * iqr)
        upper_whisker = q3 + (1.5 * iqr)
        min_value = numeric_columns[column].min()
        max_value = numeric_columns[column].max()
        mode_value = numeric_columns[column].mode()[0]

        print(f"Column Name: {column}")
        print(f"1st quartile: {q1}")
        print(f"2nd quartile: {q2}")
        print(f"3rd quartile: {q3}")
        print(f"Inter-quartile range: {iqr}")
        print(f"Lower whisker: {lower_whisker}")
        print(f"Upper whisker: {upper_whisker}")
        print(f"Min: {min_value}")
        print(f"Max: {max_value}")
        print(f"Mode: {mode_value}\n")

        # Corrected Box Plot
        plt.figure(figsize=(15, 10))
        sns.boxplot(y=numeric_columns[column], showmeans=True, notch=True, color='lightpink')
        plt.title(f'Box Plot of {column}')
        plt.show()
     
    return

print(five_num_summary(numeric_columns))
```
